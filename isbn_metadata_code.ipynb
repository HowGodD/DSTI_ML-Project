```python
import numpy as np
import pandas as pd
import os
import isbnlib
```

```python
file_path = r'C:/Users/lucyc/Documents/books.csv'
df_GoodReads = pd.read_csv(file_path, on_bad_lines='skip', index_col='bookID')
```

```python
df_GoodReads.head()
```

```python
def isbn_metadata(isbn):
    try:
        meta = isbnlib.meta(isbn)
        if meta:
            return meta
        else:
            return {'Title':'NA', 'Authors': 'NA', 'Publisher':'NA', 'Year':'NA', 'Language':'NA'}
    except Exception as e:
        return {'Title':'NA', 'Authors': 'NA', 'Publisher':'NA', 'Year':'NA', 'Language':'NA'}
    
df_isbn_meta = df_GoodReads['isbn'].apply(isbn_metadata).apply(pd.Series)
df_isbn_meta.head(25)
```

```python
df_isbn_meta.info()
```

```python
df_isbn_meta.to_csv('isbn_metadata.csv')
```
